```python
def find_data(df, new_df) :
    result = new_df.groupby(new_df['JOIN_SN']).sum()
    result = result.reset_index()
    id = df['JOIN_SN']
    
    same=[]
    for i in range(len(id)):
        a=[]
        a=result.index[result['JOIN_SN'] == id[i]].tolist() # SDB1_P_CD_USG랑 final이랑 SN 비교
        if a == []:
            continue
        else :
            same.append(a[0]) # 인덱스 리스트 생성
            
    extracted= result.loc[same]
    df = pd.merge(left = df , right = extracted, how = "left", on = "JOIN_SN" )
    df = df.fillna(0)
    return df
```
